#### 6. Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

```python
numerical = customer_df.select_dtypes(include=np.number)
numerical
```

```python
for col in numerical.columns:
    print(len(numerical[col].unique()))
```

```python
discrete_list = []
continuous_list = []
for col in numerical.columns:
    if len(numerical[col].unique()) < (customer_df.shape[0] * 0.03):
        discrete_list.append(col)
    else: 
        continuous_list.append(col)
display(discrete_list)
display(continuous_list) 
```

```python
def c_d_dif(df):
    discrete_list = []
    continuous_list = []
    for col in df.columns:
        if len(df[col].unique()) < (customer_df.shape[0] * 0.03):
            discrete_list.append(col)
        else: 
            continuous_list.append(col)
    return discrete_list,continuous_list
```

```python
discrete_list,continuous_list = c_d_dif(numerical)
```

#### 11. Did you find outliers? Comment what you will do with them.

```python
# To deal with the Outliers we can just remove them with the iqr approach as in this example: 
iqr = np.percentile(numerical['monthly_premium_auto'],75) - np.percentile(numerical['monthly_premium_auto'],25)
upper_limit = np.percentile(numerical['monthly_premium_auto'],75) + 1.5*iqr
lower_limit = np.percentile(numerical['monthly_premium_auto'],25) - 1.5*iqr
```

```python
no_outliers = numerical[(numerical['monthly_premium_auto']>lower_limit) & (numerical['monthly_premium_auto']<upper_limit)].copy()
sns.displot(no_outliers['monthly_premium_auto'])
plt.show()
```
